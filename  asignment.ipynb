#### Analyse daily return of stocks

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
#read Apple data

df1 = pd.read_csv('individual_stocks_5yr/AAPL_data.csv')

df1.head()
```

```python
df1.dtypes
```

```python
df1['date'] = pd.to_datetime(df1['date'])
```

```python
df1.dtypes
```

```python
df1 = df1.assign(day_of_week = df1['date'].dt.day_name())
```

```python
df1
```

```python
import plotly.express as px 
```

```python
fig = px.line(df1,x='day_of_week',y='close',title='AAPL')

fig.show()
```
